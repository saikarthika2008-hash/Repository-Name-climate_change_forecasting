Here from the dataset we can understand the following points:
1. The data given is from 1750 till 2024, but we will be using the data from around 1980's or 1990's because from that time period the changes are considerable percentages
2. Country column represents the name of the contry that the data belongs
3. iso_code is the code of the country
4. gdp is Gross Domestic Product it can be calculated as gdp=C+I+G+NX, where C is Consumption, I is Investment, G is Government Spending, NX is Net Exports
5. co2: Total annual carbon dioxide (CO₂) emissions from fossil fuels and industry, measured in million tonnes.
6. co2_per_capita: Average carbon dioxide emissions per person, calculated by dividing total CO₂ by the population.
7. methane: Total annual methane (CH₄) emissions, measured in million tonnes of carbon dioxide equivalents (CO₂e).
8. nitrous_oxide: Total annual nitrous oxide (N₂O) emissions, measured in million tonnes of carbon dioxide equivalents (CO₂e).
9. total_ghg: Total annual greenhouse gas emissions, including CO₂, CH₄, N₂O, and fluorinated gases, measured in million tonnes of carbon dioxide equivalents (CO₂e).

```python
df = df[df["year"] >= 1990]  # We are making the filtering process here, as mentioned in the presentaion
# We are considering the data which has year after 1990 and the rows which contain the names other than contry names
aggregates = [
    "World",
    "Asia",
    "Europe",
    "Africa",
    "North America",
    "South America",
    "European Union"
]

df = df[~df["country"].isin(aggregates)]
```

### Filtering Decisions

1. Kept records from 1990 onward because the project focuses on modern emissions trends.
2. Removed aggregate regions such as World and Asia because the analysis targets individual sovereign countries.
3. This improves consistency for future country-level modelling.

```python
import matplotlib.pyplot as plt  # Import matplotlib library for creating charts and visualizations

# Group the dataset by year and sum up CO₂ emissions across all countries
global_co2 = df.groupby("year")["co2"].sum()  

# Set the size of the chart
plt.figure(figsize=(10,5))

# Plot the global CO₂ emissions over time
plt.plot(global_co2.index, global_co2.values)

# Add a title to the chart
plt.title("Global CO₂ Emissions")

# Label the x-axis as Year
plt.xlabel("Year")

# Label the y-axis as CO₂ Emissions
plt.ylabel("CO₂ Emissions")

# Add grid lines to make the chart easier to read
plt.grid(True)

# Display the chart
plt.show()
```

Global CO₂ emissions increased steadily from 1990 onward.
Growth accelerated during the 2000s.
A noticeable dip occurs around 2020 due to COVID-19 disruptions.
Emissions recover in subsequent years.

```python
top5 = ["China", "United States", "India", "Russia", "Japan"]
```

```python
plt.figure(figsize=(12,6))  # Set the size of the chart

# Loop through each country in the top 5 emitters
for country in top5:
    # Filter the dataset for the current country
    subset = df[df["country"] == country]
    # Plot that country’s CO₂ emissions over time
    plt.plot(subset["year"], subset["co2"], label=country)

# Add a title to the chart
plt.title("CO₂ Emissions of Top 5 Emitters")

# Label the x-axis as Year
plt.xlabel("Year")

# Label the y-axis as CO₂ Emissions
plt.ylabel("CO₂ Emissions")

# Show the legend so each line is labeled by country
plt.legend()

# Display the chart
plt.show()
```

China shows the most rapid growth after 2000.
US emissions peak and gradually decline.
India shows a consistent upward trend.
Japan and Russia remain comparatively stable.

```python
df["decade"] = (df["year"] // 10) * 10  # We are categorising the years as decades so that it will be easy for the plotting partgithu
```

```python
ghg_decade = df.groupby("decade")[
    ["co2", "methane", "nitrous_oxide"]
].sum()
```

```python
ghg_decade.plot.area(figsize=(10,6))

plt.title("GHG Composition by Decade")
plt.xlabel("Decade")
plt.ylabel("Emissions")
plt.show()
```

CO₂ remains the dominant greenhouse gas.
Methane contributes a significant but smaller share.
Nitrous oxide contributes the least.
The overall volume of greenhouse gases increases across decades.
